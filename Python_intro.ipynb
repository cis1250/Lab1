### Lists
The Python List is 	**an ordered collection of elements**.


```python
# Creating a List:
# Lists are created by enclosing elements in square brackets [].

my_list = [1, 2, 3, 4, 5]
```

```python
#Adding Elements:
#You can add elements to the end of a list using the append() method.

my_list = [1, 2, 3]
my_list.append(4)
print(my_list) 
```

```python
#Inserting Elements:
#Use the insert() method to add an element at a specific index.

my_list = [1, 2, 3]
#list.insert(i, x)
#Insert an item at a given position. The first argument is the index of the element before which to insert, 
#so a.insert(0, x) inserts at the front of the list, and a.insert(len(a), x) is equivalent to a.append(x)


my_list.insert(1, 5)  
print(my_list)  # Output: [1, 5, 2, 3]
```

```python
#Removing Elements:
#Remove an element by its value using the remove() method.

my_list = [1, 2, 3, 4, 3]
my_list.remove(3)
print(my_list) 
```

## IF-ELIF-ELSE Statement


Image source: https://www.toppr.com/guides/python-guide/tutorials/python-flow-control/if-elif-else/python-if-if-else-if-elif-else-and-nested-if-statement/

```python
#if condition:
#  statements if condition is true
#elif:
#  statements if condition of ELIF is true
#else:
#  statements if condition all conditions are false

a_number = 8

if a_number % 2 == 0:
    print('{} is divisible by 2'.format(a_number))
elif a_number % 3 == 0:
    print('{} is divisible by 3'.format(a_number))
elif a_number % 5 == 0:
    print('{} is divisible by 5'.format(a_number))
else:
    print('All checks failed!')
    print('{} is not divisible by 2, 3 or 5'.format(a_number))
```

# NESTED FOR LOOPS

## FOR loops can be nested (placed inside other loops)


Image source: https://www.geeksforgeeks.org/python-nested-loops/

```python
#outer loop
for i in range(3):
    #inner loop
    for j in range(3):
        print(3*i+j,end=' ')
    print()
```

```python
rows = 5
# outer loop
for i in range(1, rows + 1):
    # inner loop
    for j in range(1, i + 1):
        print('*', end=' ')
    print()
```

```python
nterms = int(input("How many terms? "))

# first two terms
n1, n2 = 0, 1
count = 0

# check if the number of terms is valid
if nterms <= 0:
   print("Please enter a positive integer")
# if there is only one term, return n1
elif nterms == 1:
   print("Fibonacci sequence upto",nterms,":")
   print(n1)
# generate fibonacci sequence
else:
   print("Fibonacci sequence:")
   while count < nterms:
       print(n1)
       nth = n1 + n2
       # update values
       n1 = n2
       n2 = nth
       count += 1
```

```python
# Prompt the user for the number of terms
n = int(input("Enter a number: "))

# Initialize an empty list to store the triangular numbers
triangular_numbers = []

# check if the number of terms is valid
if n <= 0:
   print("Please enter a positive integer")
else:
# Generate the first n terms of the triangular number sequence
    for k in range(1, n + 1):
        T_k = k * (k + 1) // 2  # Calculate the k-th triangular number
        triangular_numbers.append(T_k)

# Print the resulting sequence
print("Triangular numbers:", triangular_numbers)
```

```python
# Get user input for book prices
book_prices = []
n = int(input("How many books are in the list? "))

for i in range(n):
    price = float(input(f"Enter the price of book {i + 1}: "))
    book_prices.append(price)

total = 0
print("\nBook prices are:")
for price in book_prices:
    print(f"${price:.2f}")
    total += price

threshold = 20
print(f"\nBooks priced above ${threshold}:")
expensive_books_found = False

for price in book_prices:
    if price > threshold:
        print(f"${price:.2f}")
        expensive_books_found = True

if not expensive_books_found:
    print("No expensive books found.")

print(f"\nTotal cost of all books: ${total:.2f}")
```
